```python
columnas_interesantes = ["Year","Province","Amount","Deposit_Area","Concept"]

datos = df_YPF[columnas_interesantes]
datos
```

```python
datos = df_YPF.copy()
# Convertir la columna 'Amount' a numérica, manejando errores
datos['Amount'] = pd.to_numeric(datos['Amount'], errors='coerce')

# Calcular el total de producción por año y provincia
total_Company = datos.groupby(["Year", "Province"])['Amount'].sum()

# Mostrar el resultado
total_Company
```

```python
# Configurar pandas para mostrar los valores en notación estándar
pd.set_option('display.float_format', '{:.2f}'.format)

# Mostrar el DataFrame resultante
print(datos)
```

```python
# Convertir la columna 'Amount' a numérica, manejando errores
datos['Amount'] = pd.to_numeric(datos['Amount'], errors='coerce')

# Calcular el total de producción por año y provincia
total_Company = datos.groupby(["Year", "Province"])['Amount'].sum()

# Mostrar el resultado
total_Company
```

```python
# Convert the 'Amount' column to numeric, handling errors
datos['Amount'] = pd.to_numeric(datos['Amount'], errors='coerce')

# Calculate the average production per year
prom_Company = datos.groupby(["Year","Province"])['Amount'].mean()

prom_Company
```

```python
# Agrupamos por Year y sumamos la columna Amount
prom_Company = datos.groupby(["Year","Province"])['Amount'].mean().reset_index()

# Mostramos el resultado
print(prom_Company)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure prom_Company is a DataFrame
prom_Company = datos.groupby(["Year","Province"])['Amount'].mean().reset_index()

# Configuración del gráfico
plt.figure(figsize=(10, 6))
sns.lineplot(data=prom_Company, x='Year', y='Amount', hue='Province', marker='o')

# Personalización
plt.title('Promedio de Amount por Año y Provincia', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Promedio de Amount', fontsize=12)
plt.legend(title='Provincia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=prom_Company, x='Year', y='Amount', hue='Province')

# Personalización
plt.title('Promedio de Amount por Año y Provincia', fontsize=16)
plt.xlabel('Año', fontsize=12)
plt.ylabel('Promedio de Amount', fontsize=12)
plt.legend(title='Provincia', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Mostrar gráfico
plt.show()
```

```python
import pandas as pd

# Supongamos que tu DataFrame se llama df
# Filtramos las filas donde el Concept es "Produccion Primaria (m3)"
produccion_primaria = df_YPF[df_YPF['Concept'] == 'Produccion Primaria (m3)']

# Convert 'Amount' column to numeric before grouping
produccion_primaria['Amount'] = pd.to_numeric(produccion_primaria['Amount'], errors='coerce')

# Agrupamos por Year y sumamos la columna Amount
total_por_año = produccion_primaria.groupby('Year')['Amount'].sum().reset_index()

# Mostramos el resultado
print(total_por_año)
```

```python
import pandas as pd

# Filtrar solo 'Produccion Primaria (m3)'
produccion_primaria = df_YPF[df_YPF['Concept'] == 'Produccion Primaria (m3)']

# Convertir 'Amount' a numérico (manejando errores)
produccion_primaria['Amount'] = pd.to_numeric(produccion_primaria['Amount'], errors='coerce')

# Calcular el PROMEDIO por año (usando mean() en lugar de sum())
promedio_por_año = produccion_primaria.groupby('Year')['Amount'].mean().reset_index()

# Renombrar la columna para claridad
promedio_por_año = promedio_por_año.rename(columns={'Amount': 'Average_Amount'})

# Mostrar el resultado
print(promedio_por_año)
```
